```python
from transformers import AutoTokenizer, AutoModel
import torch

# Load the Danish BERT model and tokenizer
tokenizer = AutoTokenizer.from_pretrained("KennethTM/bert-base-uncased-danish")
model = AutoModel.from_pretrained("KennethTM/bert-base-uncased-danish")

# Encode the input sentence
sentence = "Hvad har ejeren af en ejerlejlighed, sammen med andre ejere af lejligheder, ejendoms ret til?"
input_ids = tokenizer.encode(sentence, return_tensors="pt")  # Encode and add batch dimension

# Pass the input through the model
outputs = model(input_ids)

# Extract the CLS token representation
cls_token_final_hidden_state = outputs.last_hidden_state[:, 0, :]  # CLS token is at position 0

# Print or use the CLS token representation
print(cls_token_final_hidden_state)
```
